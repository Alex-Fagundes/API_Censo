```python
import pandas as pd

ed_basica_2023_df = pd.read_csv("microdados_ed_basica_2023.csv", sep=';', encoding='latin-1')

display(ed_basica_2023_df.head())
```

```python
ed_basica_2023_df.info(verbose=True, show_counts=True)
```

```python
#colunas = [num for num in range(260,265)]
#print(colunas)
#display(ed_basica_2023_df.iloc[[1], colunas])

util = [0,1,3,4,6,17,19,23,24,25,26,27] # 262,263,264 (tipo: presencial, semi, ead)
ed_basica_2023_filtrada_df = ed_basica_2023_df.iloc[:, util]

novo_cabecalho = {'NU_ANO_CENSO': 'Ano Censo',
                  'NO_REGIAO': 'Regiao',
                  'NO_UF': 'UF',
                  'SG_UF': 'Sigla UF',
                  'NO_MUNICIPIO': 'Municipio',
                  'NO_ENTIDADE': 'Entidade',
                  'TP_DEPENDENCIA': 'Dependencia Adm',
                 }
novo_cabecalho_endereco = {'DS_ENDERECO': 'Endereco',
                  'NU_ENDERECO': 'Numero',
                  'DS_COMPLEMENTO': 'Complemento',
                  'NO_BAIRRO': 'Bairro',
                  'CO_CEP': 'CEP',
                 }
ed_basica_2023_filtrada_df = ed_basica_2023_filtrada_df.rename(columns=novo_cabecalho)
ed_basica_2023_filtrada_df = ed_basica_2023_filtrada_df.rename(columns=novo_cabecalho_endereco)

depen_adm = {1: 'federal',
             2: 'estadual',
             3: 'municipal',
             4: 'privada',
            }
ed_basica_2023_filtrada_df['Dependencia Adm'] = ed_basica_2023_filtrada_df['Dependencia Adm'].replace(depen_adm)

display(ed_basica_2023_filtrada_df)
```
